```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
srgan = tf.keras.models.load_model("gen.h5",compile=False)
```

```python
def Predict(model, img):
    sr = model(tf.cast(tf.expand_dims(img, axis=0), tf.float32))
    sr = tf.clip_by_value(sr, 0, 255)
    sr = tf.round(sr)
    sr = tf.squeeze(tf.cast(sr, tf.uint8),axis=0)
    return sr

def load_img(path):
    img = np.array(Image.open(path))
    return img
```

```python
img = load_img("butterfly.png")
pred = Predict(srgan, img)

plt.figure(figsize=(20, 20))
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(pred)
plt.show()
```

```python
img = load_img("0963x4.png")
img = tf.image.random_crop(img, (150,150,3))
pred = Predict(srgan, img)

plt.figure(figsize=(20, 20))
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(pred)
plt.show()
```
